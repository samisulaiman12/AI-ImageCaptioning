```python
import os
import pickle
import numpy as np
import pandas as pd
import requests
from tqdm import tqdm
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical, plot_model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add
from nltk.translate.bleu_score import corpus_bleu
from tensorflow.keras.optimizers import Adam
import gc
import warnings
warnings.filterwarnings('ignore')

BASE_DIR = '/kaggle/input/flickr8k'  # Base directory for input files.
WORKING_DIR = '/kaggle/working'  # Directory for saving intermediate files.

# Load the VGG16 model
def load_vgg16_model():
    vgg_model = VGG16()  # Load pretrained VGG16 model.
    model = Model(inputs=vgg_model.inputs, outputs=vgg_model.layers[-2].output)  # Modify to output features before the last layer.
    print(model.summary())  # Print model architecture.
    return model  # Return the modified model.

# Extract image features using VGG16 model
def extract_features(model, directory):
    features = {}  # Dictionary to store features for each image.
    directory = os.path.join(BASE_DIR, 'Images')  # Directory containing images.

    for img_name in tqdm(os.listdir(directory)):  # Iterate through all images in the directory.
        img_path = directory + '/' + img_name  # Full path of the image.
        image = load_img(img_path, target_size=(224, 224))  # Load image and resize.
        image = img_to_array(image)  # Convert image to a numpy array.
        image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))  # Reshape for model input.
        image = preprocess_input(image)  # Preprocess image for VGG16.
        feature = model.predict(image, verbose=0)  # Extract features.
        image_id = img_name.split('.')[0]  # Get image ID (filename without extension).
        features[image_id] = feature  # Store the features in the dictionary.
    return features  # Return extracted features.


# Save features to pickle
def save_features_to_pickle(features, filename):
    with open(filename, 'wb') as f:
        pickle.dump(features, f)

# Load features from pickle
def load_features_from_pickle(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)

# Process and clean the captions
def clean_captions(mapping):
    for key, captions in mapping.items():  # Iterate through each image's captions.
        for i in range(len(captions)):  # Process each caption.
            caption = captions[i].lower()  # Convert to lowercase.
            caption = caption.replace('[^A-Za-z]', '').replace('\s+', ' ')  # Remove non-alphabetical characters.
            caption = 'startseq ' + " ".join([word for word in caption.split() if len(word) > 1]) + ' endseq'  # Add special tokens.
            captions[i] = caption  # Update the cleaned caption.

# Create mapping of image to captions
def create_caption_mapping(captions_doc):
    mapping = {}  # Dictionary to map images to captions.
    for line in tqdm(captions_doc.split('\n')):  # Process each line in the captions document.
        tokens = line.split(',')  # Split line by commas.
        if len(line) < 2:  # Skip invalid lines.
            continue
        image_id, caption = tokens[0], tokens[1:]  # Extract image ID and captions.
        image_id = image_id.split('.')[0]  # Remove file extension from image ID.
        caption = " ".join(caption)  # Combine caption tokens.
        if image_id not in mapping:
            mapping[image_id] = []  # Initialize list for new image ID.
        mapping[image_id].append(caption)  # Append caption to the list.
    return mapping  # Return the mapping.

# Tokenize the captions
def tokenize_captions(all_captions):
    tokenizer = Tokenizer()  # Create a tokenizer instance.
    tokenizer.fit_on_texts(all_captions)  # Fit the tokenizer on the captions to build the vocabulary.
    vocab_size = len(tokenizer.word_index) + 1  # Determine vocabulary size (add 1 for padding token).
    max_length = max(len(caption.split()) for caption in all_captions)  # Find the length of the longest caption.
    
    # Save the tokenizer to a pickle file
    with open(os.path.join(WORKING_DIR, 'tokenizer.pkl'), 'wb') as f:
        pickle.dump(tokenizer, f)
    
    return tokenizer, vocab_size, max_length  # Return the tokenizer, vocabulary size, and maximum caption length.

# Load tokenizer from pickle
def load_tokenizer_from_pickle(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)

# Create the data generator
def data_generator(data_keys, mapping, features, tokenizer, max_length, vocab_size, batch_size):
    X1, X2, y = list(), list(), list()  # Initialize empty lists for input and output data.
    n = 0  # Batch counter.
    while 1:  # Infinite loop for generating data.
        for key in data_keys:  # Iterate through each image key.
            n += 1  # Increment the batch counter.
            captions = mapping[key]  # Get the captions for the current image.
            for caption in captions:  # Process each caption.
                seq = tokenizer.texts_to_sequences([caption])[0]  # Convert caption to a sequence of integers.
                for i in range(1, len(seq)):  # Create input-output pairs from the sequence.
                    in_seq, out_seq = seq[:i], seq[i]  # Split the sequence at position i.
                    in_seq = pad_sequences([in_seq], maxlen=max_length, padding='post')[0]  # Pad the input sequence.
                    out_seq = to_categorical([out_seq], num_classes=vocab_size)[0]  # One-hot encode the output sequence.
                    X1.append(features[key][0])  # Add the image feature.
                    X2.append(in_seq)  # Add the padded input sequence.
                    y.append(out_seq)  # Add the one-hot encoded output.
            if n == batch_size:  # If the batch size is reached:
                X1, X2, y = np.array(X1), np.array(X2), np.array(y)  # Convert lists to numpy arrays.
                yield {"image": X1, "text": X2}, y  # Yield the batch as input-output pair.
                X1, X2, y = list(), list(), list()  # Reset the lists for the next batch.
                n = 0  # Reset the batch counter.
                gc.collect()  # Perform garbage collection to free up memory.


# Define the model architecture
def define_model(vocab_size, max_length):
    inputs1 = Input(shape=(4096,), name="image")  # Input layer for image features.
    fe1 = Dropout(0.4)(inputs1)  # Apply dropout to prevent overfitting.
    fe2 = Dense(256, activation='relu')(fe1)  # Dense layer to process image features.

    inputs2 = Input(shape=(max_length,), name="text")  # Input layer for text sequences.
    se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)  # Embedding layer for text sequences.
    se2 = Dropout(0.4)(se1)  # Apply dropout to text features.
    se3 = LSTM(256)(se2)  # LSTM layer to process the sequence.

    decoder1 = add([fe2, se3])  # Combine image and text features.
    decoder2 = Dense(256, activation='relu')(decoder1)  # Dense layer to refine combined features.
    outputs = Dense(vocab_size, activation='softmax')(decoder2)  # Output layer for caption prediction.

    model = Model(inputs=[inputs1, inputs2], outputs=outputs)  # Define the model with inputs and outputs.
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  # Compile the model.
    plot_model(model, show_shapes=True)  # Plot the model architecture.
    return model  # Return the compiled model.


# Train the model
def train_model(model, train_data, mapping, features, tokenizer, max_length, vocab_size, batch_size, epochs, test_data):
    steps = len(train_data) // batch_size  # Calculate the number of steps per epoch.
    accuracy_history = []  # Initialize list to store accuracy history.
    loss_history = []  # Initialize list to store loss history.

    for i in range(epochs):  # Loop over epochs.
        print(f"Epoch {i+1}/{epochs}")
        generator = data_generator(train_data, mapping, features, tokenizer, max_length, vocab_size, batch_size)  # Create generator.
        history = model.fit(generator, epochs=1, steps_per_epoch=steps, verbose=1)  # Train the model.

        accuracy_history.append(history.history['accuracy'][-1])  # Save the last epoch's accuracy.
        loss_history.append(history.history['loss'][-1])  # Save the last epoch's loss.
        
        print(f"Training Accuracy after Epoch {i+1}: {history.history['accuracy'][-1]}")
        print(f"Training Loss after Epoch {i+1}: {history.history['loss'][-1]}")
    
    model.save('best_model.keras')  # Save the trained model.

    # Plot training performance.
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(range(1, epochs + 1), accuracy_history, label='Training Accuracy')
    plt.title('Accuracy vs Epoch')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(range(1, epochs + 1), loss_history, label='Training Loss', color='red')
    plt.title('Loss vs Epoch')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

    evaluate_model(model, features, mapping, tokenizer, max_length, test_data)  # Evaluate the model on test data.


# Generate caption for an image
def generate_caption(image_name, model, tokenizer, max_length, features, mapping):
    # Extract the image ID by removing the file extension
    image_id = image_name.split('.')[0]
    # Construct the full path of the image
    img_path = os.path.join(BASE_DIR, "Images", image_name)
    # Open the image using PIL (Python Imaging Library)
    image = Image.open(img_path)
    # Predict the caption for the image
    y_pred = predict_caption(model, features[image_id], tokenizer, max_length)
    # Print the predicted caption to the console
    print('the predicted caption: ')
    print(y_pred)
    # Display the image using matplotlib
    plt.imshow(image)

# Predict the caption from the model
def predict_caption(model, image, tokenizer, max_length):
    # Start the caption with the token 'startseq'
    in_text = 'startseq'
    # Generate a sequence of words up to the max_length
    for i in range(max_length):
        # Convert the input text into a sequence of integers
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        # Pad the sequence to match the required max_length
        sequence = pad_sequences([sequence], max_length, padding='post')
        # Predict the next word using the model
        yhat = model.predict([image, sequence], verbose=0)
        # Get the index of the predicted word
        yhat = np.argmax(yhat)
        # Convert the index to a word
        word = idx_to_word(yhat, tokenizer)
        # Break if the word is not found
        if word is None:
            break
        # Append the word to the generated text
        in_text += " " + word
        # Stop if the 'endseq' token is generated
        if word == 'endseq':
            break
    # Return the generated caption
    return in_text


# Convert index to word
def idx_to_word(integer, tokenizer):
    # Search the tokenizer's word index dictionary for the matching integer
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    # Return None if the integer is not found
    return None

from nltk.translate.bleu_score import corpus_bleu
from tqdm import tqdm  # for a progress bar

# Evaluate the model using BLEU score
def evaluate_model(model, features, mapping, tokenizer, max_length, test_data):
    # Initialize lists to store actual and predicted captions
    actual, predicted = list(), list()

    # Iterate through the test set
    for key in tqdm(test_data):
        # Get actual captions
        captions = mapping[key]
        # Predict the caption for the image
        y_pred = predict_caption(model, features[key], tokenizer, max_length)
        # Split into words for BLEU scoring
        actual_captions = [caption.split() for caption in captions]  # List of lists
        y_pred = y_pred.split()  # Split predicted caption into words
        # Append to lists
        actual.append(actual_captions)
        predicted.append(y_pred)

    # Calculate BLEU scores
    print("BLEU-1: %f" % corpus_bleu(actual, predicted, weights=(1.0, 0, 0, 0)))
    print("BLEU-2: %f" % corpus_bleu(actual, predicted, weights=(0.5, 0.5, 0, 0)))


# Main execution function
def main():
    # Load the pre-trained VGG16 model
    model = load_vgg16_model()
    # Define the directory containing images
    directory = os.path.join(BASE_DIR, 'Images')
    # Extract features from images using the model
    features = extract_features(model, directory)
    # Save the extracted features to a pickle file
    save_features_to_pickle(features, os.path.join(WORKING_DIR, 'features.pkl'))
    
    # Read captions from the file
    with open(os.path.join(BASE_DIR, 'captions.txt'), 'r') as f:
        next(f)  # Skip the header line
        captions_doc = f.read()

    # Create a mapping of image IDs to their captions
    mapping = create_caption_mapping(captions_doc)
    # Clean the captions (e.g., remove special characters)
    clean_captions(mapping)
    
    # Tokenize all captions into a vocabulary
    all_captions = [caption for captions in mapping.values() for caption in captions]
    tokenizer, vocab_size, max_length = tokenize_captions(all_captions)
    
    # Split data into training and testing sets
    image_ids = list(mapping.keys())
    split = int(len(image_ids) * 0.95)
    train_data = image_ids[:split]
    test_data = image_ids[split:]
    
    # Define the captioning model
    model = define_model(vocab_size, max_length)
    # Train the model on the training set
    train_model(model, train_data, mapping, features, tokenizer, max_length, vocab_size, batch_size=32, epochs=50, test_data=test_data)

# Execute the main function if the script is run directly
if __name__ == "__main__":
    main()
```
